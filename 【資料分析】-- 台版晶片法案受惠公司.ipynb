為鞏固臺灣產業國際供應鏈關鍵地位，經濟部擬訂的「產業創新條例第10條之2及第72條」修正草案(又稱臺版晶片法案)，於1/7立法院三讀通過修正案。本法適用對象不限產業別，只要在國內進行技術創新且位居國際供應鏈關鍵地位之公司，均得申請適用；其中，研發費用、研發密度、有效稅率訂為適用優惠減稅的三大必備門檻；符合適用要件者，得享有前瞻創新研發投資抵減，當年度抵減率為25%、購置用於先進製程之設備投資抵減，當年度抵減率為5%，支出金額無上限；施行期間自112年1月1日至118年12月31日止。目前相關子法尚未定調，消息指出，研發經費門檻或將訂在50億至100億元，研發密度則會落在5%至7%之間；對此財政部及經濟部兩部正為此角力，深怕門檻訂太高變成台積電獨享，又怕門檻訂得太低變成家家有獎，失去鼓勵意義。以下將以臺灣全體上市公司的研發費用及研發密度數據來分析該法案。

## 1. Download data & Retrieving company industry

Industry name referring from https://www.twse.com.tw/zh/statistics/statisticsList?type=05&subType=225

在數據取得與整理方面，首先我們根據臺灣證交所的資料，將臺股整體上市公司作產業分類；再藉由個別上市公司的研究經費(Capex)及營業收入淨額(Revenue)，前者除以後者，算出研發密度(Capex_Density)。經過數據整理後，以2021年財報數據來分析。

```python
import pandas as pd
import numpy as np
```

```python
from finlab.data import Data
data = Data()

T3100營業收入淨額 = data.get("營業收入合計")
T3356研究發展費 = data.get("研究發展費用").fillna(data.get("研究發展費用合計")).fillna(data.get("委託研究費"))
```

```python
# Drop the one with no Revenue & Capex e.g. ETF
# shifting the dataset because financial release date is the next quarter
# Unit: $ Thousand TWD
T3356研究發展費 = T3356研究發展費.dropna(how='all', axis=1).shift(-1).resample('Y').sum()
T3100營業收入淨額 = T3100營業收入淨額.dropna(how='all', axis=1).shift(-1).resample('Y').sum()
```

```python
T3100營業收入淨額
```

```python
# Retrieving industry type
ticker = pd.read_excel('電子業ticker.xls', sheet_name='2022Q2')
ticker = ticker.iloc[5:, 0:2].dropna(how='any')
ticker.columns = ['ticker', 'name']
ticker = ticker.reset_index(drop=True)

industry_name = []  # to fill "XX類"
refer_name = []     # to record "XX類", and fill it in industry_name
row_num = []        # to drop "XX類" row

for i in ticker.index: 
    if len(ticker['ticker'][i]) == 2: 
        row_num.append(i)
        refer_name.append(ticker['name'][i])
    else: 
        industry_name.append(refer_name[-1])
        
ticker = ticker.drop(row_num)
ticker = ticker.reset_index(drop=True)
ticker['industry'] = pd.DataFrame(industry_name)
ticker = ticker.set_index('ticker')

ticker
```

## 2. Data Cleaning

```python
def all_data(year): 
    
    capex = T3356研究發展費.loc[year].transpose()
    capex.columns = ['capex']
    revenue = T3100營業收入淨額.loc[year].transpose()
    revenue.columns = ['revenue']
    
    # combine ticker + revenue + capex
    result = pd.concat([ticker, capex.reindex(ticker.index), revenue.reindex(ticker.index)], axis=1)
    # calculate "capex daensity" = capex/revenue
    result['capex_density'] = result['capex']/result['revenue'] 
    
    # EDA：inf取代為nan；nan取代為0
    result.replace([np.inf, -np.inf], np.nan, inplace=True)
    result.replace(np.nan, 0, inplace=True)
    
    return result
```

```python
all_data('2020')
```

## 3. Calculate beneficiaries of the Act

```python
# Create a function as choosing the year of data
def beneficiaries(result, capex_thrd, density_thrd): 
    
    # beneficiary companies 
    condition1 = result['capex']>capex_thrd              # capex > 5000000 (current)
    condition2 = result['capex_density']>density_thrd    # density > 0.05 (current)
    matching_company = result[condition1&condition2]
    print(matching_company['name'].count())
    return matching_company
    
    #"Total Numebers of Beneficiaries:" + 
```

## 4. Data Analysis

```python
test1 = all_data('2021')
beneficiaries(test1, 10000000, 0.07)
```

```python
test2 = all_data('2021')
beneficiaries(test2, 2000000, 0.05)
```

#### (2) which one is harder? condition 1 or 2. Using data from year of 2021

若分別將兩個條件來比較，條件「研究經費達50億臺幣」將只篩選出26間業者；條件「研究密度達5%」將會篩選出184間業者，可見前者的條件較不易達標，多數科技但非半導體業者(如：電腦及週邊設備業、電子零組件業、通信網路業、光電業)雖然研發密度高，但因為公司規模、研究經費不夠高，而被踢出去名單。

```python
data = all_data('2021')
beneficiaries(data, 5000000, 0)
beneficiaries(data, 0, 0.05)
# Apparently, condition1 is harder, only filter out 26 companies
```

```python
beneficiaries(data, 2000000, 0.04)
```

## 5. Conclusion

雖然該法案適用對象不限產業別，部分業者雖然研發密度已達門檻，但是受阻於研發經費門檻相對高，該法案恐只有規模較大的半導體業者可以受惠。另外，本次門檻可以明顯發現傳統產業難以達到門檻，由於不同產業受到本身產業體質不同，研發密度無法與半導體業相比擬，政府或斟酌多方考量產業本身的特性，再擬定該法令的門檻值，如此才能避免該法案成為半導體業專用。金融市場方面，留意該法案相關子法訂定在接下來成為話題，屆時最終受惠的臺股上市公司名單將出爐。
